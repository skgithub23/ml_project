#Plotting Relation between price range and battery power

```python
plt.figure(figsize=(4,3))
sea.barplot(x='price_range',y = 'battery_power', data= data)
plt.show()
```

Plotting Relation between price range and pixel height/ width

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sea.barplot(x='price_range',y='px_height',data=data, palette= 'Blues')
plt.subplot(1,2,2)
sea.barplot(x = 'price_range', y = 'px_width', data= data, palette= 'Greens')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sea.barplot(x='price_range',y='ram',data= data)
plt.show()
```

```python
corr = data.corr()
corr.shape
```

```python
plt.figure(figsize=(14,14))
sea.heatmap(corr, cbar=False, square= True, fmt='.2%', annot=True, cmap='Greens')
plt.show()
```
